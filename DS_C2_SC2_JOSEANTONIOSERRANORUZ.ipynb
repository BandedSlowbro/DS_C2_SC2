```python
import numpy as np
import pandas as pd 
```

a. Extraer la información del archivo.

```python
vacunacion = pd.read_csv('country_vaccinations.csv')
vacunacion.head(3)
```

b. Mostrar la estructura y tipos de datos de cada columna para identificar qué operaciones puedes realizar con cada una de ellas, asegurándote que las columnas con fechas sean del tipo datetime64. 

```python
vacunacion['date']=pd.to_datetime(vacunacion['date'])
vacunacion.info()
```

c. Determinar la cantidad de vacunas aplicadas de cada compañía (con base en cómo lo reporta cada país en la columna vaccines). 

```python
#Creando una nueva columna 'canVaccines' almacenamos la cantidad de vacunas aplicadas de cada compañía 
#tomando en cuenta que el método de separación es una coma ', ' 
### se creó un dataframe para no añadir la nueva columna al daraframe original. 

df_canVaccines= vacunacion.copy()
df_canVaccines['canVaccines'] = df_canVaccines['vaccines'].str.split(', ').str.len()
df_canVaccines['canVaccines']
```

d. Obtener la cantidad de vacunas aplicadas en todo el mundo. 

```python
#Para obtener la cantidad de vacunas aplicadas en todo el mundo podemos hacer una sumatoria de toda la columna 
# de vacunaciones diarias 

vaccapplied_total = vacunacion['daily_vaccinations'].sum()

print("La cantidad de vacunas aplicadas en todo el mundo es:",vaccapplied_total)
```

e. Calcular el promedio de vacunas aplicadas por país. 

```python
# Para calcular el promedio de vacunas por país, agrupamos por país y calculamos el promedio: 
vacunacion[['country','daily_vaccinations']].groupby('country').mean()
```

f. Determinar la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo. 

```python
# Primero se crea un dataframe en donde se seleccionan todas las líneas con la fecha buscada
# posteriormente sumamos la columna daily_vaccinations para encontrar todas las vacunaciones de ese dataframe:

vacunacion_290121 = vacunacion[vacunacion['date'] == '29-01-2021']

# Calcular la cantidad total de vacunas aplicadas ese día en todo el mundo
total_vacunas = np.sum(vacunacion_290121['daily_vaccinations'])

print(f"La cantidad total de vacunas aplicadas el día 29/01/21 en todo el mundo fue de {total_vacunas} vacunas.")
```

g. Crear un dataframe nuevo denominado conDiferencias que contenga los datos originales y una columna derivada (diferencias) con las diferencias de aplicación entre las columnas daily_vaccionations y daily_vaccionations_raw.  

```python
conDiferencias = vacunacion.assign(diferencias=vacunacion['daily_vaccinations'] - vacunacion['daily_vaccinations_raw'])
conDiferencias
```

h. Obtener el período de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua. 

```python
#Obtenemos la fecha mínima y la fecha máxima y procedemos con el cáculo: 

print('Fecha del primer reporte: ', vacunacion['date'].min())
print('Fecha del último reporte: ', vacunacion['date'].max())

periodo = vacunacion['date'].max() - vacunacion['date'].min()

print("El período de tiempo entre la fecha más antigua y la más reciente es:", periodo)
```

i. Crear un dataframe nuevo denominado conCantidad que contenga los datos originales y una columna derivada (canVac) con la cantidad de vacunas utilizadas cada día (usar la columna vaccines y separar por el carácter /). 

```python
# Utilicé el mismo planteamiento del paso c. pero de acuerdo a esta intrucción, separar por el caracter "/" en lugar de "," 
conCantidad= vacunacion.copy()
conCantidad['canVac'] = conCantidad['vaccines'].str.split('/').str.len()
conCantidad
```

j. Generar un dataframe denominado antes20 con todos los registros que se hayan realizado antes del 20 de diciembre de 2020. 

```python
# Con loc se pueden obtener todos los registros donde se cumpla la condición especificada
antes20 = vacunacion.loc[vacunacion['date'] < '20-12-2020']
antes20
```

k. Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna Pfizer. 

```python
#Dado que las celdas de vaccines contienen varios caracteres, podemos buscar la palabra "Pfizer", filtrando todas las filas que la contengan

pfizer = vacunacion[vacunacion['vaccines'].str.contains('Pfizer')]
pfizer
```

l. Almacenar los dataframes generados (conDiferencias, conCantidad, antes20 y pfizer) en un archivo de Excel denominado resultadosReto.xlsx, donde cada dataframe ocupe una hoja diferente. Se recomienda ver la documentación de pd.ExcelWriter. 

```python
#De acuerdo a la documentación, podemos usar de base el ejemplo dado en "To write to separate sheets in a single file:"

with pd.ExcelWriter("resultadosReto.xlsx") as writer:
    conDiferencias.to_excel(writer, sheet_name="conDiferencias")  
    conCantidad.to_excel(writer, sheet_name="conCantidad")  
    antes20.to_excel(writer, sheet_name="antes20")
    pfizer.to_excel(writer, sheet_name="pfizer")
```
